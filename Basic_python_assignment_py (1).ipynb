Question 3. Compare and contrast mutable and immutable objects in Python with
            examples.

Answer 3.  In Python, mutable and immutable objects differ in whether their
           values can be changed after creation.

           Mutable Objects:
                          Can be changed after creation.
                          Examples: Lists, dictionaries, sets.
      
Example:
       my_list = [1, 2, 3]
my_list[0] = 10  # Modifying the first element
print(my_list)  # Output: [10, 2, 3]

           Immutable Objects:
                            Cannot be changed after creation.
                            Examples: Strings, tuples, integers.

Example:
      my_string = "hello"
# my_string[0] = "H"  # This would raise an error as strings are immutable
my_string = "Hello"  # You can reassign but not modify the original string
print(my_string)  # Output: "Hello".

Question 4.  Discuss the different types of operators in Python and provide
             examples of how they are used.

Answer 4.   1.  Arithmetic Operators: Perform mathematical operations.
              + (Addition): 5 + 3 # 8
              - (Subtraction): 5 - 3 # 2
              * (Multiplication): 5 * 3 # 15
              / (Division): 5 / 3 # 1.6667
              % (Modulus): 5 % 3 # 2
              ** (Exponent): 2 ** 3 # 8
              // (Floor Division): 5 // 3 # 1
              
            2. Comparison Operators: Compare values and return True or False.
              == (Equal to): 5 == 3 # False
              != (Not equal to): 5 != 3 # True
              > (Greater than): 5 > 3 # True
              < (Less than): 5 < 3 # False
              >= (Greater than or equal to): 5 >= 3 # True
              <= (Less than or equal to): 5 <= 3 # False

            3. Logical Operators: Combine conditional statements.
              and: True and False # False
              or: True or False # True
              not: not True # False

            4.  Assignment Operators: Assign values to variables.
                =: x = 5
                +=: x += 2 # Same as x = x + 2
                -=: x -= 1 # Same as x = x - 1

Question 5.  Explain the concept of type casting in Python with examples.

Answer 5.  Type casting in Python is the process of converting one data type to
           another. It’s helpful when you need to perform operations between incompatible types or display data in a different format.

           Example.

            b = 2.5 # float

Question 7.  Describe the different types of loops in Python and their use  
             cases with examples.

Answer 7.  Python has two primary types of loops: for loop and while loop. Both
           are used to execute a block of code repeatedly, but they are suited for different scenarios based on the task or condition.

           1. for Loop
           The for loop is used to iterate over a sequence of elements (such as lists, tuples, strings, or ranges). It runs the block of code once for each item in the sequence.

           Syntax:
           
           for variable in sequence:
           # Code to execute

           Use Cases:
           Iterating over a list or range.
           Performing actions on each item in a collection.
           Example:

           # Loop through a list of numbers
           numbers = [1, 2, 3, 4, 5]

           for num in numbers:
           print(num)
           Output:
           1
           2
           3
           4
           5
           In this example, the loop iterates over each number in the list and prints it.

          Example with range():

          # Using range() to generate numbers from 0 to 4
          for i in range(5):
               print(i)
               Output:
                     0
                     1
                     2
                     3
                     4


            2. while Loop
               The while loop repeatedly executes a block of code as long as the specified condition is True.

               Syntax:

                while condition:
                # Code to execute
                   Use Cases:
                     Repeating tasks until a condition changes (like waiting for user input).
                     Running a loop when the number of iterations isn’t known beforehand.
                     
                     Example:
                       
                     # Print numbers from 1 to 5 using while loop
                    i = 1
                    while i <= 5:
                        print(i)
                        i += 1  # Increment i by 1

                      Output:
                            1
                            2
                            3
                            4
                            5
                            In this example, the loop continues until i is greater than 5.

            3. break and continue Statements
               break: Exits the loop entirely when a certain condition is met.
               continue: Skips the current iteration and moves to the next one.
                         Example with break:

                       # Break out of the loop when number is 3
                       for num in range(1, 6):
                           if num == 3:
                           break
                               print(num)
                           Output:
                                  1
                                  2
                           Example with continue:

                           # Skip number 3 and continue the loop
                           for num in range(1, 6):
                               if num == 3:
                                  continue
                               print(num)
                           Output
                           1
                           2
                           4
                           5
              4. else with Loops
                 Python allows an else block to be used with both for and while loops. The else block is executed after the loop finishes normally (i.e., not interrupted by a break).

                        Example with else:

                        for num in range(5):
                            print(num)
                        else:
                            print("Loop finished!")
                        Output:

                              0
                              1
                              2
                              3
                              4
                        Loop finished!


                        Use Case Summary:
                        for loop: Best when you know the number of iterations (like iterating through a list or range).
                        while loop: Useful when you want to repeat until a condition is met and you don’t know the number of iterations upfront.

                        break and continue: Help to control the flow of the loop by exiting or skipping specific iterations.
